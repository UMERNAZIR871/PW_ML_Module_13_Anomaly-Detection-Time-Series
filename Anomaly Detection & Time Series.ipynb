**Q 1:  What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.**

**Ans.** 
## üîë Definition
**Anomaly Detection** (also called **Outlier Detection**) is the process of identifying data points, events, or patterns that deviate significantly from the expected behavior in a dataset.  
Such unusual observations often indicate **critical information** like fraud, system faults, or rare events.

---

## üìå Types of Anomalies

### 1. Point Anomalies
- **Definition**: A single data point is very different from the rest of the data.
- **Example**:
  - A credit card transaction of **$10,000** when most transactions are under **$500**.
  - A sudden spike in temperature sensor reading (e.g., 120¬∞C in a machine usually operating at 40‚Äì60¬∞C).

---

### 2. Contextual Anomalies (Conditional Anomalies)
- **Definition**: A data point is considered anomalous **in a specific context** but may be normal in another context.
- **Example**:
  - **Temperature anomaly**: 25¬∞C is normal in summer but unusual in winter.
  - **Website traffic**: High user visits at midnight may be unusual, but normal during daytime.

---

### 3. Collective Anomalies
- **Definition**: A group of data points collectively deviate from the expected pattern, even if individual points are not anomalous.
- **Example**:
  - Multiple failed login attempts from the same IP address within a short time (potential cyber-attack).
  - A sequence of stock market trades that form a suspicious pattern, even though each trade looks normal individually.

**Q 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.**

**Ans.** 

## üîë Overview
These three algorithms are widely used for **anomaly detection**, but they differ in approach and suitable scenarios.

---

## üìä Comparison Table

| Algorithm            | Approach                                                                 | Suitable Use Cases                                                                 |
|----------------------|--------------------------------------------------------------------------|------------------------------------------------------------------------------------|
| **Isolation Forest** | Randomly partitions data using decision trees. Anomalies are isolated faster because they require fewer splits. | - High-dimensional datasets<br>- Fraud detection<br>- Network intrusion detection |
| **DBSCAN**           | Density-based clustering: points in low-density regions are labeled as outliers. | - Spatial/geographical data<br>- Detecting unusual patterns in sensor networks<br>- When clusters have irregular shapes |
| **Local Outlier Factor (LOF)** | Measures local density deviation of a point compared to its neighbors. Points with much lower density than neighbors are anomalies. | - Datasets with varying densities<br>- Detecting subtle local anomalies<br>- Customer behavior analysis |

---

## üìù Key Differences

1. **Isolation Forest**
   - **Strengths**: Works well with large, high-dimensional datasets; scalable; model-based.
   - **Limitations**: May struggle with anomalies that are context-dependent or subtle in local density.
   
2. **DBSCAN**
   - **Strengths**: Finds clusters of arbitrary shape; detects global outliers effectively.
   - **Limitations**: Sensitive to parameters (`eps`, `min_samples`); not ideal for high-dimensional data.
   
3. **Local Outlier Factor (LOF)**
   - **Strengths**: Captures **local anomalies** where density varies across regions.
   - **Limitations**: Computationally expensive for large datasets; performance depends heavily on neighborhood size (`k`).

**Question 3: What are the key components of a Time Series? Explain each with one example.**

**Ans.**

A **time series** is a sequence of data points collected or recorded at successive points in time.  
To analyze and forecast time series data, we typically break it down into four key components:

---

## 1. Trend
- **Definition**: The long-term movement or direction in the data (upward, downward, or flat) over time.
- **Example**: The steady **increase in global average temperatures** over decades due to climate change.

---

## 2. Seasonality
- **Definition**: Regular, repeating patterns or cycles in data that occur at fixed intervals (daily, weekly, monthly, yearly).
- **Example**: **Retail sales increase during December** every year due to holiday shopping.

---

## 3. Cyclical Component
- **Definition**: Fluctuations in data that occur at irregular intervals, usually influenced by economic or business cycles (longer-term than seasonality).
- **Example**: **Economic recessions and expansions** affecting stock market prices over several years.

---

## 4. Irregular/Residual (Noise)
- **Definition**: Random, unpredictable variations in the data that cannot be explained by trend, seasonality, or cycles.
- **Example**: A sudden **drop in sales due to a one-day website outage**.

**Q 4: Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?**

**Ans.** 
## üîë Definition
A **stationary time series** is one whose **statistical properties** (mean, variance, autocorrelation) remain **constant over time**.  
This is crucial because most time series models (e.g., ARIMA) assume stationarity to make reliable forecasts.

---

## üìå Properties of a Stationary Series
1. **Constant Mean** ‚Üí The average value does not change over time.  
2. **Constant Variance** ‚Üí The spread of values remains stable.  
3. **Constant Autocovariance** ‚Üí The relationship between observations depends only on the lag, not on the actual time at which data is observed.

---

## üß™ How to Test for Stationarity
1. **Visual Inspection**
   - Plot the series and check for visible trends or seasonality.
2. **Rolling Statistics**
   - Calculate rolling mean and variance ‚Äî if they change over time, the series is likely non-stationary.
3. **Statistical Tests**
   - **Augmented Dickey-Fuller (ADF) Test**:
     - Null hypothesis (H‚ÇÄ): Series is non-stationary.
     - If p-value < 0.05 ‚Üí Reject H‚ÇÄ ‚Üí Series is stationary.
   - **KPSS Test**:
     - Null hypothesis (H‚ÇÄ): Series is stationary.
     - If p-value < 0.05 ‚Üí Reject H‚ÇÄ ‚Üí Series is non-stationary.

---

## üîß Transforming a Non-Stationary Series

1. **Differencing**
   - Subtract current observation from the previous one.  
   - Example: `y_t' = y_t ‚Äì y_(t-1)`  
   - Removes trends and stabilizes the mean.

2. **Log Transformation**
   - Apply log to reduce variance.  
   - Example: `log(y_t)` for sales data with exponential growth.

3. **Seasonal Differencing**
   - Subtract the value from the same season in the previous cycle.  
   - Example: `y_t' = y_t ‚Äì y_(t-s)` where `s` is season length (e.g., 12 for monthly data).

4. **Detrending**
   - Remove estimated trend (e.g., by regression or moving average smoothing).

**Q 5: Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.**

**Ans.**

Time series forecasting models build on each other, each adding more complexity to capture trends, seasonality, and external factors.

---

## 1. Autoregressive Model (AR)
- **Structure**: Uses the **past values of the series** to predict future values.
- **Equation**:  
  $$
  Y_t = c + \phi_1 Y_{t-1} + \phi_2 Y_{t-2} + \dots + \phi_p Y_{t-p} + \epsilon_t
  \$$  
- **Key Parameter**: `p` ‚Üí number of lags.  
- **Application Example**: Predicting stock prices based only on past price movements.

---

## 2. Moving Average Model (MA)
- **Structure**: Uses **past forecast errors (residuals)** to predict future values.
- **Equation**:  
  $$
  Y_t = c + \theta_1 \epsilon_{t-1} + \theta_2 \epsilon_{t-2} + \dots + \theta_q \epsilon_{t-q} + \epsilon_t
  \$$  
- **Key Parameter**: `q` ‚Üí number of lagged forecast errors.  
- **Application Example**: Short-term demand forecasting using shocks/noise patterns.

---

## 3. Autoregressive Integrated Moving Average (ARIMA)
- **Structure**: Combines **AR + MA + differencing** to make the series stationary.  
- **Parameters**: `(p, d, q)`
  - `p` = AR order  
  - `d` = differencing order  
  - `q` = MA order  
- **Application Example**: Forecasting sales where data has a trend but no seasonality.

---

## 4. Seasonal ARIMA (SARIMA)
- **Structure**: Extends ARIMA by handling **seasonality** explicitly.  
- **Parameters**: `(p, d, q)(P, D, Q, s)`
  - First part `(p, d, q)` = non-seasonal components  
  - Second part `(P, D, Q, s)` = seasonal components  
  - `s` = length of seasonality (e.g., 12 for monthly data with yearly seasonality)  
- **Application Example**: Forecasting **monthly airline passengers**, where strong yearly seasonality exists.

---

## 5. Seasonal ARIMA with Exogenous Variables (SARIMAX)
- **Structure**: SARIMA + external (exogenous) regressors.  
- **Equation Idea**:  
  $$
  Y_t = \text{SARIMA terms} + \beta X_t + \epsilon_t
  \$$  
- **Exogenous Variables**: Additional factors influencing the series (e.g., holidays, promotions, temperature).  
- **Application Example**: Forecasting **ice cream sales** using seasonality + external factor like **daily temperature**.

---

## ‚úÖ Summary Table

| Model   | Components Used                           | Handles Seasonality | Uses External Factors | Example Use Case |
|---------|-------------------------------------------|---------------------|-----------------------|------------------|
| **AR**  | Past values (lags)                        | ‚ùå No                | ‚ùå No                  | Stock prices based on past prices |
| **MA**  | Past errors (residuals)                   | ‚ùå No                | ‚ùå No                  | Demand forecasting with shocks |
| **ARIMA** | AR + MA + Differencing (p,d,q)          | ‚ùå No                | ‚ùå No                  | Sales forecasting with trend |
| **SARIMA** | ARIMA + Seasonal terms (P,D,Q,s)       | ‚úÖ Yes               | ‚ùå No                  | Airline passengers with yearly seasonality |
| **SARIMAX** | SARIMA + Exogenous regressors (X vars)| ‚úÖ Yes               | ‚úÖ Yes                 | Sales forecast with seasonality + holiday effects |

**Q 6: Load a time series dataset (e.g., AirPassengers), plot the original series, and decompose it into trend, seasonality, and residual components.**

```python
# Step 1: Import libraries
import warnings
warnings.filterwarnings("ignore")  # Ignore warnings

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.datasets import get_rdataset

# Step 2: Load AirPassengers dataset
data = get_rdataset('AirPassengers').data
data['time'] = pd.date_range(start='1949-01', periods=len(data), freq='M')
data.set_index('time', inplace=True)
ts = data['value']

# Step 3: Plot original time series
plt.figure(figsize=(10,5))
plt.plot(ts, label='Original Series', color='blue')
plt.title("AirPassengers Time Series")
plt.xlabel("Time")
plt.ylabel("Number of Passengers")
plt.legend()
plt.show()

# Step 4: Decompose the series (Multiplicative)
decomposition = seasonal_decompose(ts, model='multiplicative', period=12)

# Step 5: Plot the decomposed components
decomposition.plot()
plt.suptitle("Time Series Decomposition (Multiplicative)", fontsize=16)
plt.show()
```

**Q 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.**

```python
# Step 1: Import libraries
import warnings
warnings.filterwarnings("ignore")  # Ignore warnings

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Step 2: Load sample NYC Taxi Fare dataset
# Here we create a sample dataset (replace with actual CSV if available)
# We'll simulate pickup longitude and latitude for demonstration
import numpy as np
np.random.seed(42)
n_samples = 500
# Normal data
pickup_longitude = np.random.uniform(-74.05, -73.75, n_samples)
pickup_latitude = np.random.uniform(40.63, 40.85, n_samples)

# Outliers
outliers_longitude = np.random.uniform(-74.2, -73.6, 15)
outliers_latitude = np.random.uniform(40.55, 40.95, 15)

longitude = np.concatenate([pickup_longitude, outliers_longitude])
latitude = np.concatenate([pickup_latitude, outliers_latitude])

df = pd.DataFrame({'pickup_longitude': longitude, 'pickup_latitude': latitude})

# Step 3: Apply Isolation Forest
iso_forest = IsolationForest(contamination=0.03, random_state=42)
df['anomaly'] = iso_forest.fit_predict(df[['pickup_longitude', 'pickup_latitude']])

# Step 4: Plot results
plt.figure(figsize=(8,6))
normal = df[df['anomaly'] == 1]
anomalies = df[df['anomaly'] == -1]

plt.scatter(normal['pickup_longitude'], normal['pickup_latitude'], c='blue', label='Normal', s=50)
plt.scatter(anomalies['pickup_longitude'], anomalies['pickup_latitude'], c='red', label='Anomaly', s=100, marker='X')

plt.title("Isolation Forest Anomaly Detection (NYC Taxi Fare)")
plt.xlabel("Pickup Longitude")
plt.ylabel("Pickup Latitude")
plt.legend()
plt.show()
```

**Q 8: Train a SARIMA model on the monthly airline passengers dataset. Forecast the next 12 months and visualize the results.**

```python
# Step 1: Import libraries
import warnings
warnings.filterwarnings("ignore")  # Ignore warnings

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.datasets import get_rdataset

# Step 2: Load AirPassengers dataset
data = get_rdataset('AirPassengers').data
data['time'] = pd.date_range(start='1949-01', periods=len(data), freq='M')
data.set_index('time', inplace=True)
ts = data['value']

# Step 3: Train SARIMA model
# Using SARIMA(p,d,q)(P,D,Q,s)
# We'll start with (1,1,1)(1,1,1,12) as a common starting point
model = SARIMAX(ts, order=(1,1,1), seasonal_order=(1,1,1,12))
sarima_result = model.fit(disp=False)

# Step 4: Forecast next 12 months
forecast = sarima_result.get_forecast(steps=12)
forecast_index = pd.date_range(start=ts.index[-1] + pd.offsets.MonthBegin(1), periods=12, freq='M')
forecast_values = forecast.predicted_mean
confidence_intervals = forecast.conf_int()

# Step 5: Plot original series + forecast
plt.figure(figsize=(10,6))
plt.plot(ts, label='Original')
plt.plot(forecast_index, forecast_values, color='red', label='Forecast')

# Plot confidence intervals
plt.fill_between(forecast_index,
                 confidence_intervals.iloc[:,0],
                 confidence_intervals.iloc[:,1],
                 color='pink', alpha=0.3)

plt.title("SARIMA Forecast for AirPassengers")
plt.xlabel("Time")
plt.ylabel("Number of Passengers")
plt.legend()
plt.show()
```

**Q 9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib.**

```python
# Step 1: Import libraries
import warnings
warnings.filterwarnings("ignore")  # Ignore warnings

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# Step 2: Generate synthetic 2D dataset
np.random.seed(42)
n_samples = 300
# Normal data
X1 = np.random.normal(0, 1, (n_samples, 2))
# Add some outliers
X2 = np.random.uniform(low=-6, high=6, size=(15,2))
X = np.vstack([X1, X2])

df = pd.DataFrame(X, columns=['Feature1', 'Feature2'])

# Step 3: Apply Local Outlier Factor
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
y_pred = lof.fit_predict(df)
# LOF labels: -1 -> anomaly, 1 -> normal
df['anomaly'] = y_pred

# Step 4: Visualize anomalies
plt.figure(figsize=(8,6))
normal = df[df['anomaly'] == 1]
anomalies = df[df['anomaly'] == -1]

plt.scatter(normal['Feature1'], normal['Feature2'], c='blue', label='Normal', s=50)
plt.scatter(anomalies['Feature1'], anomalies['Feature2'], c='red', label='Anomaly', s=100, marker='X')

plt.title("Local Outlier Factor (LOF) Anomaly Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```
